```python
# Librerias, archivos y configuración

import warnings
import pandas as pd
import os
import mysql.connector as conn
from sqlalchemy import create_engine

import funciones

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
%%time

# Se hace una lista con los archivos de original_data para poder llamar a la función

archivos = os.listdir('../original_data')
archivos.sort(reverse=True)

dfs = [e for e in archivos]

# Se llama a la función que va a unir todos los csv

total = funciones.unir(dfs)
```

```python
# Se crean los df estaciones, magnitudes y unidades_medida para su registro

lugares = {4: 'Pza. de España',
8: 'Escuelas Aguirre', 
11: 'Av. Ramón y Cajal',
16: 'Arturo Soria',
17: 'Villaverde Alto',
18: 'Calle Farolillo',
24: 'Casa de Campo',
27: 'Barajas',
35: 'Pza. del Carmen',
36: 'Moratalaz',
38: 'Cuatro Caminos',
39: 'Barrio del Pilar',
40: 'Vallecas',
47: 'Méndez Álvaro',
48: 'Po. Castellana',
49: 'Retiro',
50: 'Pza. Castilla',
54: 'Ensanche Vallecas',
55: 'Urb. Embajada (Barajas)',
56: 'Plaza Elíptica',
57: 'Sanchinarro',
58: 'El Pardo',
59: 'Parque Juan Carlos I',
60: 'Tres Olivos'}

estaciones = pd.DataFrame.from_dict(lugares, orient='index', columns=['estacion'])
estaciones.reset_index(inplace=True)
estaciones.columns = ['estaciones_id', 'estacion']


magnitudes = {1: 'Dióxido de Azufre',
6: 'Monóxido de Carbono', 
7: 'Monóxido de Nitrógeno',
8: 'Dióxido de Nitrógeno',
12: 'Óxidos de Nitrógeno',
9: 'Partículas < 2.5 um',
10: 'Partículas < 10 um',
14: 'Ozono',
20: 'Tolueno',
30: 'Benceno',
35: 'Etilbenceno',
}

magnitudes = pd.DataFrame.from_dict(magnitudes, orient='index', columns=['magnitud'])
magnitudes.reset_index(inplace=True)
magnitudes.columns = ['magnitudes_id', 'magnitud']


unidades = {1: 'ug/m3',2: 'mg/m3', }

unidades_medida = pd.DataFrame.from_dict(unidades, orient='index', columns=['unidad'])
unidades_medida.reset_index(inplace=True)
unidades_medida.columns = ['unidades_id', 'unidad']
```

```python
# MySQL connect

str_conn = 'mysql+pymysql://root:password@localhost:3306/contaminacion_madrid'
cursor = create_engine(str_conn)

# Importar los datos a SQL

try:
    estaciones.to_sql(name='Estaciones', con=cursor, if_exists='append', index=False)
    magnitudes.to_sql(name='Magnitudes', con=cursor, if_exists='append', index=False)
    unidades_medida.to_sql(name='unidades_medida', con=cursor, if_exists='append', index=False)
    total.to_sql(name='Datos_anteriores', con=cursor, if_exists='replace', index=False)

except: total.to_sql(name='Datos_anteriores', con=cursor, if_exists='replace', index=False)
```
